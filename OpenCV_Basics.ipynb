# GrayScale Conversion

```python
import cv2
image  = cv2.imread("i2.jpg")
if(image is not None):
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    cv2.imshow("Grayscale Image",gray)
    cv2.imshow("Original image",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not loaded")
```

# Resizing Images

```python
import cv2
image = cv2.imread("i2.jpg")
if(image is not None):
    print("Image loaded")
    resized = cv2.resize(image,(800,300))   # cv2.resize(src,dsize ->(width,height),fx,fy,interpolation) #interpolation for pixles resizing ie. to improve resolution
    cv2.imshow("Original Image",image)
    cv2.imshow("Resized image",resized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.imwrite("resized.png",resized)
else:
    print("Image not loaded")
    
```

# Image Rotation

```python
import cv2
# M=cv2.getRotationMatrix2D(center,angle,scale)
# rotated_image = cv2.warpAffine(image,M,(width,height))

image = cv2.imread("dino.jpg")

if(image is None):
    print("Could not load image...")
else:
    (h,w) = image.shape[:2]
    center = (w//2,h//2)
    M=cv2.getRotationMatrix2D(center,140,1.0)
    rotated = cv2.warpAffine(image,M,(w,h)) #w,h can be change
    cv2.imshow("Original",image)
    cv2.imshow("Rotated image: ",rotated)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

# Flipping

```python
import cv2 
image = cv2.imread("dino.jpg")
# flipped= cv2.flip(image,flipcode)   1-> horizontal 0->vertical -1->both
if(image is not None):
    flipped_horizontal = cv2.flip(image,1)
    flipped_vertical = cv2.flip(image,0)
    flipped_both = cv2.flip(image,-1)

    cv2.imshow("Original image",image)
    cv2.imshow("Horizonatal Flipped image",flipped_horizontal)
    cv2.imshow("Vertically flipped image",flipped_vertical)
    cv2.imshow("Both sided flipped image",flipped_both)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Not loaded image")
```

# Drawing line

```python
import cv2
image = cv2.imread("i2.jpg")
#cv2.line(image,point1,point2,color,thickness)
if(image is not None):
    print("Image loaded successfully!!!")
    print("Height",image.shape[0])
    print("Width",image.shape[1])
    pt1=(100,400)
    pt2 = (100,900)
    color=(0,0,255)
    thickness=4
    cv2.line(image,pt1,pt2,color,thickness)
    cv2.imshow("Image is ",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not loaded")
```

# Drawing Rectangle

```python
# cv2.rectangle() -> cv2.rectangle(img,pt1,pt2,color,thickness) pt1 is top left corner and pt2 is bottom right corner of the rectangle
import cv2
image = cv2.imread("i2.jpg")
if(image is None):
    print("Image not loaded")
else:
    print("Height is:",image.shape[0])
    print("Width:",image.shape[1])
    pt1=(50,400)
    pt2=(400,700)
    color=(255,0,0)
    thickness=4  
    cv2.rectangle(image,pt1,pt2,color,thickness)  # if thickness is in negative then it fill the rectangle
    cv2.imshow("Image is: ",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
