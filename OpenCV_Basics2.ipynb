# Drawing Circle

```python
import cv2
image = cv2.imread("i2.jpg")
# cv2.cirlce(image,centre,radius,color,thickness)
if(image is None):
    print("Image not loaded")
else:
    print("Image loaded successfully!!")
    cv2.circle(image,(210,150),50,(255,0,0),3)
    cv2.imshow("Drawing Circle",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

# Adding Text - cv2.putText()

```python
import cv2 
image =cv2.imread("i2.jpg")
#cv2.putText(img,text,org,font,fontScale,color,thickness)
# org is bottom left corner of the text 
#fontscale is 1.0 then remain same 2.0 -> double 
if(image is None):
    print("Image not loaded!!")
else:
    print("Image loaded successfully!!")
    h,w = image.shape[:2]
    print("Height: ",h)
    print("Width: ",w)
    # Get text size
    text = input("Enter you want to write: ")
    fontscale = float(input("Enter font scale: "))
    thickness = int(input("Enter thickness : "))
    cv2.putText(image,text,(300,400),cv2.FONT_HERSHEY_SIMPLEX,fontscale,(255,255,0),thickness)
    cv2.imshow("Image is:",image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
