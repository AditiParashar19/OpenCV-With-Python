# Opencv-grayscale-converter

```python
import cv2
img = input("Enter image name: ")
image = cv2.imread(img)
if(image is not None):
    ch = int(input("Enter your choice\n1->Convert to grayscale\n 2-> Save the gray scale image\n3->Only show the image"))
    if(ch==1):
        gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        print("Successfully converted to grayscale image!!!")
    elif(ch==2):
        gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        cv2.imwrite("gray_scale.png",gray)
        print("Image saved successfully!!")
    elif(ch==3):
        gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        cv2.imshow("Original image is :",image)
        cv2.imshow("GrayScale image: ",gray)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print("Wrong choice")
else:
    print("Image not found")
```
